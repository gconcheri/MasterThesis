```python
def nonzeropairs(mat):
    nonzero_pairs = []
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            if mat[i, j] != 0:
                nonzero_pairs.append([i, j])

    return nonzero_pairs
```
